###3.2. Colonna "data_creazione"

Altro elemento da gestire è il passaggio della colonna `data_creazione` da *stringhe* a `int`.

Per trasformare la colonna `data_creazione` da `object` a `Int64`, bisogna innanzitutto gestire i valori che contengono intervalli (es. `1240 | 1250`) e quelli che non sono numeri. Estraggo il primo anno da eventuali intervalli e converto il tutto in un tipo intero che supporta i valori mancanti (`pd.NA`).

```python
# Definisco una funzione chiamata "converti_data_creazione" per gestire i valori nulli e i valori che contengono intervalli

def converti_data_creazione(data):
    if pd.isna(data):
        return pd.NA
    str_data = str(data)
    if ' | ' in str_data:
        # Prende il primo anno nell'intervallo
        anno_str = str_data.split(' | ')[0]
    else:
        anno_str = str_data

    try:
        return int(anno_str)
    except ValueError:
        # Gestisce i casi in cui la stringa non è un numero valido
        return pd.NA

# Applico la funzione alla colonna 'data_creazione'
df['data_creazione'] = df['data_creazione'].apply(converti_data_creazione)

# Converto la colonna al tipo Int64 (intero con supporto per NA)
df['data_creazione'] = df['data_creazione'].astype('Int64')

# Verifico il nuovo tipo di dato della colonna
display(df.info())
```

Ora la colonna `data_creazione` è di tipo `Int64`, e i valori non validi sono stati convertiti in `<NA>`. Visualizzo le prime e le ultime righe per confermare.

```python
display(df.head())
display(df.tail())
```

### 4.2. Il genere e l'arte

Quanti sono gli artisti maschi e quante le femmine in questo DataSet?

Per effettuare questo calcolo, innanzitutto creo un nuovo DataFrame che contenga solo i valori della colonna `id` e `artisti`.

Considero anche le righe con valori nulli, corrispondenti ad artisti anonimi o i cui nomi per qualche altro motivo non sono stati riportati nel DataSet.





```python
# Uso .copy() per esplicitare a Pandas che sto lavorando su una copia, e non sul DataFrame originale

df_artisti = df[["id","artisti"]].copy()
df_artisti
```

Devo trovare il modo di estrarre dalla colonna `artisti` il genere dell'artista.  Noto che il genere è fra parentesi.

Itero sulle righe del DataFrame "**df_artisti**" per ricavare il genere dell'artista da ogni riga. Se la colonna `artisti` presenta un valore nullo, faccio sì che il genere risulti `non specificato`.

```python
for index, row in df_artisti.iterrows():
    # Noto che alcuni artisti non hanno il nome popolato, per questo eseguo questa operazione solo su chi ha il nome
    if pd.isna(row['artisti']) == False:
      nome = row['artisti']
      genere = nome.split("(")[1].split(")")[0]
      df_artisti.loc[index, 'artisti'] = genere
    else:
      df_artisti.loc[index, 'artisti'] = "non specificato"

# Cerco tutti i valori non conformi e noto che la riga 2440 presenta tra parentesi un valore non conforme: "(pittore 1861-1948)"
for index, row in df_artisti.iterrows():
    if row['artisti'] != "non specificato" and row['artisti'] != "maschio" and row['artisti'] != "femmina":
      display(row)
```

```python
# Lo sistemo assegnando il valore "maschio"
df_artisti.loc[2440, 'artisti'] = "maschio"

# Cerco eventuali altri valori non conformi
for index, row in df_artisti.iterrows():
    if row['artisti'] != "non specificato" and row['artisti'] != "maschio" and row['artisti'] != "femmina":
      display(row)

#Sarebbe corretto anche rinominare la colonna appena ripopolata
df_artisti = df_artisti.rename(columns={'artisti': 'genere_artista'})

display(df_artisti)
```

C'è un legame tra il genere femminile e le dimensioni delle opere prodotte? Idealmente, le opere di dimensioni maggiori richiedono una committenza importante. Studiando i dati del dataset, si può stabilire se anche le donne ricevevano commissioni importanti.
- Calcolo l'area media di tutte le opere.
- Quanto misura l'area delle opere delle artiste femmine?
- Confronto: l'area media degli artisti maschi è in linea con quella delle artiste femmine? Se sì, possiamo dire con un certo grado di approssimazione che venivano impiegati entrambi in media per commissioni della stessa importanza.

```python
# Aggiungo al dataframe originale una colonna che indichi per tutti l'area delle opere
df["area_opera"] = df["altezza"] * df["larghezza"]

# Visualizzo l'area delle artiste femmine creando un nuovo DataFrame chiamato "df_femmine"
df_femmine = df[df['genere_artista'] == 'femmina'].copy()
df_femmine
```

```python
# Calcolo la media dell'area delle opere di artiste donne creando un nuovo DataFrame chiamato "area_media_femmine"
area_media_femmine = df_femmine['area_opera'].mean()

# Creo un nuovo DataFrame contenente solo gli artisti maschi e calcolo l'area media delle loro opere
df_maschi = df[df['genere_artista'] == 'maschio'].copy()
area_media_maschi = df_maschi['area_opera'].mean()

# Mostro i risultati
display(area_media_femmine), display(area_media_maschi)
```

Visualizzo questi risultati in un grafico a barre.

```python
ax = df.groupby("genere_artista")["area_opera"].mean().plot.barh(
    color=['#AB0000', '#C1D5DC', '#C1D5DC'],
    figsize=(12,5)
    )

ax.set_facecolor('#f0f0f0')
ax.grid(color='white', linewidth=1.5)
ax.set_axisbelow(True)
for bordo in ax.spines.values():
    bordo.set_visible(False)
colore_testo = '#444444'
ax.set_title("Dimensioni medie delle opere di artisti maschi e femmine",
             loc="left",
             pad=20,
             fontsize=16,
             color=colore_testo)
ax.tick_params(colors=colore_testo, labelsize=11)
ax.xaxis.label.set_color(colore_testo)
ax.set_xlabel("Dimensioni opere (cm²)", fontsize=12, labelpad=10)
ax.set_ylabel("Genere artista", fontsize=12, labelpad=10)
```

Il risultato è molto interessante: sembrerebbe che, in media, le opere delle artiste donna siano più grandi rispetto a quelle degli uomini.

Potrebbero esserci pochi dati che influenzano il risultato, considerato il ridotto numero di opere di artiste di genere femminile.

Sicuramente bisogna tener conto della dimensione ristretta del campione di opere di artiste di genere femminile rispetto a quello dei colleghi di genere maschile, però questo risultato costituisce comunque un dato inatteso e interessante da approfondire.

###5.3. Il legame tra "generi" e opere di grandi dimensioni
Per assicurarmi che il risultato di queste indagini non risulti falsato da poche opere di dimensioni eccezionalmente grandi, indago con un ultimo quesito se le opere di genere `arte religiosa` e `ritratto` siano anche quelle di maggiori dimensioni.

Lo faccio visualizzando in un grafico a dispersione la quantità di opere di artiste femminili divise per genere artistico in rapporto alla dimensione dell'area di ognuna.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_femmine_exploded = df_femmine.assign(generi=df_femmine['generi'].str.split('; ')).explode('generi').reset_index(drop=True)

# Creo un grafico a dispersione per visualizzare l'area delle opere per genere
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_femmine_exploded, x='generi', y='area_opera', hue='generi', s=500, alpha=0.8, legend=False)

plt.title('Area delle opere di artiste donne per tutti i generi', fontsize=20)
plt.xlabel('Genere artistico', fontsize=17)
plt.ylabel('Area opera (cm²)', fontsize=17, labelpad=20)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Come si nota dal grafico, la maggior parte delle opere di grandi dimensioni appartengono al genere `arte religiosa` e `ritratto`, ad eccezione di due valori **outlier** appartenenti ai generi `nudo artistico` e `arte di genere` (che potrebbero far riferimentto ad un'unica opera, dato che ho applicato `.explode()` sulla colonna dei generi).
